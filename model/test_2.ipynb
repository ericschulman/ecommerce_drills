```julia
import Pkg
using Optim
using PyPlot
using LinearAlgebra


Pkg.add("FFTW")

using FFTW
```

```julia
#https://stackoverflow.com/questions/56030394/how-to-visualize-fft-of-a-signal-in-julia
# Number of points 
N = 10

t = 1:N
print(t)
```

```julia
# signal 
signal = repeat([0,1,2,3,2],2)

# Fourier Transform of it 
F = fft(signal)

# plots 
plot(collect(t), signal, label="time")
plot(collect(t), abs.(F),label="frequency")

inv = F[:]
inv= inv .* (abs.(F) .< 5*ones(N)) #filter high frequencies
print(inv)
inv = ifft(inv)

plot(collect(t), inv,label="frequency 2")
legend()
show()

#ifft(F)
```
